## Задача №69. Решение в группах
- Изобразить гистограмму по flipper_length_mm с оттенком height_group. 
- Сделать анализ

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("penguins.csv.")
def long(data):
    if data >= 42:
        return "high"
    elif 42 > data > 35:
        return "middle"
    elif data == None:
        return
    return "low"
df["height_group"] = df["bill_length_mm"].apply(long)
df.head()
```

```python
sns.displot(x = "flipper_length_mm", hue="height_group", data = df, aspect=2, height=5, kind="kde")
```
