```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("penguins.csv")
df.head()
```

```python
sns.scatterplot(data=df, x="island", y="body_mass_g", hue="body_mass_g")
```

```python
sns.scatterplot(data=df, x="species", y="bill_length_mm", hue="body_mass_g")
```

```python
sns.scatterplot(data=df, x="species", y="body_mass_g", hue="sex")
```

```python
cols = ['bill_depth_mm', 'bill_length_mm', 'flipper_length_mm', 'body_mass_g']
g = sns.PairGrid(df[cols])
g.map(sns.scatterplot)
```
