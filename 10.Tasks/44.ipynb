В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего
из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без
get_dummies?
```

```

````python
import random
import pandas as pd
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data
````

Вид с помощью **get_dummies()** 

````python
pd.get_dummies(data)
````

Вид **one hot** без dummy

````python
roboman = pd.DataFrame({'Man':lst, 'Robot':lst})
roboman['Man'] = data['whoAmI'].map({'human': 1, 'robot':0})
roboman['Robot'] = data['whoAmI'].map({'human': 0, 'robot':1})
roboman
````
